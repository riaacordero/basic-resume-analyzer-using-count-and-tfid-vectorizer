```python
import docx2txt
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
job_description = docx2txt.process("./files/job_description.docx")
resume = docx2txt.process("./files/resume.docx")
```

```python
print(resume)
```

```python
print(job_description)
```

```python
content = [job_description, resume]
```

```python
cv = CountVectorizer()
cv_matrix = cv.fit_transform(content)
cv_similarity_matrix = cosine_similarity(cv_matrix)
print("Resume matches by (Count Vectorizer): ",str(round(cv_similarity_matrix[1][0]*100))+"%")
```

```python
tf = TfidfVectorizer()
tf_matrix = tf.fit_transform(content)
similarity_matrix = cosine_similarity(tf_matrix)
print("Resume matches by (TFid Vectorizer): ",str(round(similarity_matrix[1][0]*100))+"%")

# Count Vectorizer gives equal weightage to all the words in the document in which it includes stopwords like "the", "is", "and" etc.
# TFIDF vectorizer ignores the stopwords and gives more weight to rare words that can be more informative. Thus, it is more accurate.
```
